```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.ketquadientoan.com/tat-ca-ky-xo-so-power-655.html?datef=30-08-2023&datet=11-05-2024"
#url = "https://www.ketquadientoan.com/tat-ca-ky-xo-so-mega-6-45.html?datef=01-01-2022&datet=26-04-2024"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
raw_data = soup.find_all('td')
print(raw_data)
```

```python
# Tạo danh sách chứa các số từ mỗi dòng
numbers_list = []

# Lặp qua từng thẻ <td> trong raw_data
for td in raw_data:
    # Tìm tất cả các thẻ <span> trong <td>
    spans = td.find_all("span", class_="home-mini-whiteball")
    # Lấy nội dung của từng thẻ <span> và chuyển đổi thành số nguyên
    numbers = [int(span.get_text()) for span in spans]
    # Chỉ lấy 6 số đầu tiên và thêm chúng vào numbers_list nếu có ít nhất một số trong mảng
    if any(numbers):
        numbers_list.append(numbers[:6])

print(numbers_list)
```

```python
# Tạo DataFrame từ danh sách hai chiều
df = pd.DataFrame(numbers_list)

# Đường dẫn đến file Excel
excel_file_path = "output.xlsx"

# Xuất DataFrame ra file Excel
df.to_excel(excel_file_path, index=False, header=False)

print("Xuất dữ liệu thành công vào file Excel:", excel_file_path)
```

```python
import re
# Lấy ngày tháng năm từ raw_data, ra mảng ngày tháng năm, vd: ['01-01-2022', '02-01-2022', '03-01-2022', '04-01-2022', '05-01-2022', '06-01-2022']
date_list = []

# Lặp qua từng phần tử trong raw_data
for td in raw_data:
    # Tìm chuỗi chứa ngày tháng năm
    date_match = re.search(r'\d{2}/\d{2}/\d{4}', td.get_text())
    if date_match:
        date_list.append(date_match.group())

print(date_list)
```

```python
#xuất dữ liệu date_list ra file date_list.xlsx
df_date = pd.DataFrame(date_list)
df_date.to_excel("date_list.xlsx", index=False, header=False)
print("Xuất dữ liệu ngày tháng năm thành công vào file Excel: date_list.xlsx")
```

```python
import pandas as pd

# Đọc dữ liệu từ các tệp Excel
df_date = pd.read_excel("date_list.xlsx", header=None)
df_numbers = pd.read_excel("output.xlsx", header=None)

# Đặt tên cho các cột
df_date.columns = ["Date"]
df_numbers.columns = ["1st_number", "2nd_number", "3rd_number", "4th_number", "5th_number", "6th_number"]

# Ghép 2 DataFrame
df_merged = pd.concat([df_date, df_numbers], axis=1)

# Đặt tên cho các cột trong DataFrame ghép
df_merged.columns = ["Date", "1st_number", "2nd_number", "3rd_number", "4th_number", "5th_number", "6th_number"]

# Lưu DataFrame vào tệp Excel
df_merged.to_excel("previous_data.xlsx", index=False)

print("Ghép cột ngày tháng năm và cột số thành công vào file Excel: previous_data.xlsx")
```
